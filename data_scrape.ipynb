```
import requests
from bs4 import BeautifulSoup


url = 'http://kenyu1234.php.xdomain.jp/record.php'
r = requests.get(url)
html_content = r.text
soup = BeautifulSoup(html_content, 'lxml')

links = soup.find_all('a')
```

```
links
```

```
links = links[5:]
```

```
s =  str("<a href='person.php?name=18'>建部和歌夫★</a>")
s=s[25:-4]
s
```

```
nl = []
for link in links:
  link = str(link)
  link = link.replace('★','')
  link = link[25:-4]
  link = link.replace('>','')
 # link = link.replace('<','')
  nl.append(link.split('"')+[0])
  
nl
```

```
base_url = 'http://kenyu1234.php.xdomain.jp/persony.php?name=@&je=0'
birth_url = 'http://kenyu1234.php.xdomain.jp/person.php?name=@'

for entry in nl:
  num = entry[0]
  entry[0] = base_url.replace('@',num)
  entry[2] = birth_url.replace('@',num)
  
nl  
```

```
new = pd.DataFrame()

for i in nl[:5]:
  data = pd.read_html(i[0])
  data[0]['name'] = i[1]
  print(data[0])
  new = pd.concat([new,data[0]])

  
new
```

```
summary = pd.DataFrame()
matches = pd.DataFrame()

for i in nl:
  data = pd.read_html(i[0])
  data[0]['name'] = i[1]
  data[1]['name'] = i[1]
  print(data[0])
  summary = pd.concat([summary,data[0]])
  matches = pd.concat([matches,data[1]])
```
